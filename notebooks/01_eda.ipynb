```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title("Tenure vs Churn")
plt.show()
```

### ðŸ“Œ Insight: Tenure vs Churn

- Customers with low tenure are more likely to churn.
- Long-term customers are more loyal.
- Tenure is a strong predictor of churn.


```python
plt.figure(figsize=(8,5))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Contract Type vs Churn")
plt.xticks(rotation=20)
plt.show()
```

### ðŸ“Œ Insight: Contract Type vs Churn

- Customers with month-to-month contracts have the highest churn rate.
- One-year contract users show significantly lower churn.
- Two-year contract users are the most loyal with very minimal churn.
- Longer contract duration improves customer retention.
- Contract type is a strong predictor for churn analysis.


```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

### ðŸ“Œ Insight: Monthly Charges vs Churn

- Customers who churn tend to have higher monthly charges compared to loyal customers.
- The median monthly charge for churned users is higher than for non-churned users.
- High service costs may contribute to customer dissatisfaction.
- MonthlyCharges is an important numerical feature for predicting churn.


```python
plt.figure(figsize=(8,5))
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title("Internet Service vs Churn")
plt.show()
```

### ðŸ“Œ Insight: Internet Service vs Churn

- Customers using Fiber Optic internet services have the highest churn rate.
- DSL users show comparatively lower churn.
- Customers without internet service are the most loyal.
- High churn among fiber users may be due to higher costs or service expectations.
- Internet service type is an important categorical feature for churn prediction.


```python
plt.figure(figsize=(8,5))
sns.countplot(x='gender', hue='Churn', data=df)
plt.title("Gender vs Churn")
plt.show()
```

### ðŸ“Œ Insight: Gender vs Churn

- The churn rates for male and female customers are almost similar.
- There is no significant difference in churn behavior based on gender.
- Gender does not appear to be a strong predictor of customer churn.
- Other factors such as contract type, tenure, and monthly charges have a greater impact on churn.
